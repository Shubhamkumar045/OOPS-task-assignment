```python
Q1-- Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
     and average_of_vehicle.
Answer--class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_speed = average_speed

# Creating instances of the Vehicle class
car = Vehicle("Car", 200, 100)
bike = Vehicle("Bike", 160, 60)

# Accessing the instance variables
print("Car:")
print("Name:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average Speed:", car.average_speed)

print("\nBike:")
print("Name:", bike.name_of_vehicle)
print("Max Speed:", bike.max_speed)
print("Average Speed:", bike.average_speed)

Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
    Create a method named seating_capacity which takes capacity as an argument and returns the name of
    the vehicle and its seating capacity.
Answer--class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_speed = average_speed

class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} people."

# Creating an instance of the Car class
car = Car("Car", 200, 100)

# Accessing attributes and calling the method
print("Name:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average Speed:", car.average_speed)
print(car.seating_capacity(5))
#output
Name: Car
Max Speed: 200
Average Speed: 100
Car has a seating capacity of 5 people.

Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
Answer--class Parent1:
    def method1(self):
        print("Method 1 from Parent 1")

class Parent2:
    def method2(self):
        print("Method 2 from Parent 2")

# Child class inheriting from both Parent1 and Parent2
class Child(Parent1, Parent2):
    def child_method(self):
        print("Child Method")

# Creating an instance of the Child class
child_obj = Child()

# Calling methods from both parent classes and child class
child_obj.method1()  # Call method from Parent1
child_obj.method2()  # Call method from Parent2
child_obj.child_method()  # Call method from Child
#output
Method 1 from Parent 1
Method 2 from Parent 2
Child Method

Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
    class.
Answer--Getter and setter methods are used to control the access and modification of attributes (instance variables) of a class.
        They provide a way to ensure encapsulation and data integrity by allowing controlled access to attribute values.

        Getter Method: A getter method is used to retrieve the value of an attribute.
        It ensures that you can access the value of an attribute through a method, which can be useful for adding 
        additional logic or validation before returning the value.

        Setter Method: A setter method is used to set the value of an attribute. 
        It allows you to control how a value is assigned to an attribute,
        which can be helpful for adding validation or performing certain actions when setting a value.
class Student:
    def __init__(self, name, age):
        self._name = name  # Private attribute with single underscore
        self._age = age    # Private attribute with single underscore

    # Getter method for 'name'
    def get_name(self):
        return self._name

    # Setter method for 'name'
    def set_name(self, name):
        if isinstance(name, str):
            self._name = name
        else:
            print("Name must be a string.")

    # Getter method for 'age'
    def get_age(self):
        return self._age

    # Setter method for 'age'
    def set_age(self, age):
        if isinstance(age, int) and age >= 0:
            self._age = age
        else:
            print("Age must be a non-negative integer.")

# Creating an instance of the Student class
student = Student("Alice", 20)

# Using getter and setter methods
print("Original Name:", student.get_name())
student.set_name("Bob")
print("Updated Name:", student.get_name())

print("Original Age:", student.get_age())
student.set_age(25)
print("Updated Age:", student.get_age())

Q5.What is method overriding in python? Write a python code to demonstrate method overriding.
Answer--
```
